```python
import pandas as pd
import numpy as np
import requests
import matplotlib.pyplot as plt
import json
%matplotlib inline
```

### Analyzing and Visualizating

```python
df_tweets_clean = pd.read_csv('twitter_archive_master.csv')
```

Let's look is there any dependencies between retweets count and favorites count (it's clearly should be, but I plan to ensure by looking at plot below). 

```python
#let's look for dependencies between likes and retweets. We can see it clearly
plt.figure(figsize=(10,6))
plt.scatter(df_tweets_clean.retweet_count,df_tweets_clean.favorite_count);
plt.title('Dependencies of retweets and favoritrs');
plt.xlabel('Retweets count');
plt.ylabel('Favorites count');
```

```python
#Lets evaluate distribution of ratings.
plt.figure(figsize=(14,8))
df_tweets_clean.groupby('rating_numerator')['rating_denominator'].count().plot(kind='bar')
plt.xlabel('Ratings');
plt.ylabel('Number of occurrences');
plt.title('Distribution of ratings');
```

```python
df_tweets_clean.rating_numerator.value_counts()
```

```python
df_tweets_clean.groupby('rating_numerator')['favorite_count'].mean().plot(kind='bar',figsize = (15,10),title = 'The average number of likes in each category of dog ratings');
plt.xlabel("Dog ratings");
plt.ylabel('Average favorites');
```

```python
df_tweets_clean.groupby('rating_numerator')['retweet_count'].mean().plot(kind='bar',figsize = (15,10),title = 'The average number of retweets in each category of dog ratings');
plt.xlabel("Dog ratings");
plt.ylabel('Average retweets');
```

```python
plt.figure(figsize=(15,10))
plt.scatter(df_tweets_clean.rating_numerator,df_tweets_clean.favorite_count);
plt.xticks(range(19));
plt.title('Ratings vs favorite count');
plt.xlabel('Rating');
plt.ylabel('Favorites');
```

```python
df_tweets_clean.groupby('stages')['rating_numerator'].mean().plot(kind='bar',figsize=(15,10),title='Average ratings for stages');
plt.xlabel('Stages');
plt.ylabel('Ratings');
```

```python
df_tweets_clean.groupby('stages')['favorite_count'].mean().plot(kind='bar',figsize=(15,10),title='Average likes for stages');
plt.xlabel('Stages');
plt.ylabel('Likes');
```

```python
df_tweets_clean.groupby('stages')['retweet_count'].mean().plot(kind='bar',figsize=(15,10),title="Average retweet's for stages");
plt.xlabel('Stages');
plt.ylabel('Likes');
```

```python
highest_ratings = df_tweets_clean[df_tweets_clean.conf_coef>0.7].groupby('breed')['rating_numerator'].mean().sort_values(ascending=False)[:15]
highest_ratings.plot(kind = 'bar',figsize = (18,10),title = 'Most rated breeds');
plt.ylabel('Average rating');
```

```python
highest_ratings
```

```python
most_favorites = df_tweets_clean[df_tweets_clean.conf_coef>0.7].groupby('breed')['favorite_count'].mean().sort_values(ascending=False)[:10]
most_favorites.plot(kind = 'bar',figsize = (18,10),title = 'Most liked breeds');
plt.ylabel('Average likes');
```

```python
most_favorites
```

```python
most_retweeted = df_tweets_clean[df_tweets_clean.conf_coef>0.7].groupby('breed')['retweet_count'].mean().sort_values(ascending=False)[:10]
most_retweeted.plot(kind = 'bar',figsize = (18,10),title = 'Most retweeted breeds');
plt.ylabel('Average retweets');
```

```python
most_retweeted
```

```python
most_common = df_tweets_clean[df_tweets_clean.conf_coef>0.7].groupby('breed')['breed'].count().sort_values(ascending=False)
labels = np.array(most_common.index)
labels[20:] = ''
most_common.plot(kind = 'pie',figsize = (17,17),title = 'Most common breed',labels =labels );
```
